```python
import torch
import cv2
import numpy as np

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5:v6.2', 'yolov5s', pretrained=True)
```

```python
video_path = r'D:\women_safety\vid.MP4'
cap = cv2.VideoCapture(video_path)
```

```python

    def detect_violence(current_positions):
        # Simple heuristic: Check if any two persons are too close to each other
        for i in range(len(current_positions)):
            for j in range(i + 1, len(current_positions)):
                distance = np.linalg.norm(np.array(current_positions[i]) - np.array(current_positions[j]))
                if distance < 50:  # Adjust this threshold
                    return True
        return False
```

```python
  def analyze_motion(prev_positions, current_positions):
        # Calculate movement by comparing positions between frames
        for prev, curr in zip(prev_positions, current_positions):
            if np.linalg.norm(np.array(prev) - np.array(curr)) > 30:  # Threshold for aggressive movement
                return True
        return False
```

```python
if not cap.isOpened():
    print("Error: Could not open video.")
else:
    print("Video file is open successfully")
    total_frames = 0
    violent_frames = 0
    violence_detected = False

    # Start reading and processing the video frames
    while True:  # Correct indentation here
        ret, frame = cap.read()
        if not ret:
            break

        total_frames += 1
           # Analyze the frame
        results = model(frame)

        current_positions = []
        for *box, conf, cls in results.xyxy[0]:
            x1, y1, x2, y2 = map(int, box)
            label = f'{model.names[int(cls)]} {conf:.2f}'
            if model.names[int(cls)] == 'person':
                current_positions.append((x1, y1, x2, y2))
                cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        # Check for violence detection (close proximity + aggressive movement)
        if detect_violence(current_positions) or analyze_motion(prev_positions, current_positions):
            violent_frames += 1
            violence_detected = True  # Set flag to True
            cv2.putText(frame, "Violence Detected!", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 255), 2)

        # Update previous positions
        prev_positions = current_positions

        # Display the message if violence is detected
        if violence_detected:
            cv2.putText(frame, "Violence Detected!", (frame.shape[1] - 300, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Display the frame with detections
        cv2.imshow('Frame', frame)

        if cv2.waitKey(1) & 0xFF == 27:
            break

    # After processing the video, calculate the percentage of violent frames
    if total_frames > 0:
        violence_percentage = (violent_frames / total_frames) * 100
        print(f"Violence detected in {violence_percentage:.2f}% of the video frames.")
        if violence_percentage >= 50:
            print("Violence Detected - Over 50% of the video is marked as violent.")

cap.release()
cv2.destroyAllWindows()
```
